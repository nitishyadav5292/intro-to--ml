```python
import pandas as pd
import numpy as np
df = pd.read_csv('MR data - Compiled Data Set.csv')
df.head(5)
```

```python
df = df.drop('Title', axis = 1)
df = df.drop('Text', axis=1)
df.head(10)
```

```python
df = df.fillna(0)
df.head(10)
df.columns = [i for i in range(0,len(df.columns))]
df.head(10)
for i in range(0, len(df)):
    if len(str(df.iloc[i:8]).split(','))>2:
        print(i, df.iloc[i,:])
        dicn = {0:0,"Property Disputes":1, "Family Dispute":2, "Petty Quarrels":3, "Money Disputes":4, "Personal Vendetta":5, "Love Affairs":6, "Casteism":7, "Unknown reasons":8}
df[1] = df[1].apply(lambda x: dicn[x])
```

```python
df.head(40)
```

```python
df[11]=0
df[12]=0
```

```python
for i in range(0,len(df)):
    #print(df.iloc[i,8])
    if ',' in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
    elif int(df.iloc[i,8])>100:
        df.iloc[i,11]=int(df.iloc[i,8])%100
        df.iloc[i,8]=int(df.iloc[i,8])//100
```
