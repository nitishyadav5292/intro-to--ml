```python
#since the number of rows are very large than 2, so we can easily delete 2 rows for uniformity in data
df = df.dropna()
df.isnull().sum()
print('Dataframe dimensions:', df.shape)
df.head()
```

```python
df.drop('TAIL_NUM',axis=1,inplace=True)
df.head()
```

```python
df = df[df['Dew Point'].map(lambda x: x.isascii())]
df.shape
```

```python
X = df.drop(['TAXI_OUT'],axis = 1)
Y = df.TAXI_OUT
print(X.shape)
print(Y.shape)
X.head()
```
